```python
from ultralytics import YOLO

def train_yolo():
    # Load YOLOv8 pretrained model
    model = YOLO("yolov8n.pt")

    # Train the model
    model.train(
        data="/kaggle/working/data.yaml",
        epochs=50,
        imgsz=640,
        batch=16,
        workers=4,
        project="traffic_sign_detection",
        name="yolov8_traffic_signs",
        exist_ok=True
    )

if __name__ == "__main__":
    train_yolo()
```

```python
from ultralytics import YOLO

def evaluate_model():
    model = YOLO("traffic_sign_detection/yolov8_traffic_signs/weights/best.pt")
    
    metrics = model.val()
    
    print("Evaluation Results:")
    print(metrics)

if __name__ == "__main__":
    evaluate_model()
```

```python
from ultralytics import YOLO
import cv2
import yt_dlp
import os

# Load YOLO model
model = YOLO(r"C:\\private\\AI\\assingment\\arch tech\\Traffic Signs Recognition\\best (1).pt")

# YouTube video URL
youtube_url = "https://www.youtube.com/watch?v=wqctLW0Hb_0"

# Download path
video_file = "youtube_video.mp4"

# Download YouTube video
ydl_opts = {
    'outtmpl': video_file,
    'format': 'mp4'
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([youtube_url])

# Open video
cap = cv2.VideoCapture(video_file)

# Output video
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(
    "detected_youtube_output.mp4",
    fourcc,
    30,
    (int(cap.get(3)), int(cap.get(4)))
)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame, conf=0.4)
    annotated_frame = results[0].plot()

    out.write(annotated_frame)
    cv2.imshow("Traffic Sign Detection - YouTube", annotated_frame)

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

# Load trained YOLO model
model = YOLO(r"C:\\private\\AI\\assingment\\arch tech\\Traffic Signs Recognition\\best (1).pt")
youtube_url = "https://www.youtube.com/watch?v=wqctLW0Hb_0"
# Load video instead of webcam
video_path = "youtube_video.mp4"
cap = cv2.VideoCapture(video_path)

# Check video opened
if not cap.isOpened():
    print(" Video file open nahi ho rahi")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print(" Video finished")
        break

    # YOLO inference
    results = model(frame, conf=0.4)

    # Draw detections
    annotated_frame = results[0].plot()

    # Show result
    cv2.imshow("Traffic Sign Detection (Video)", annotated_frame)

    # Control playback speed (optional)
    if cv2.waitKey(25) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
